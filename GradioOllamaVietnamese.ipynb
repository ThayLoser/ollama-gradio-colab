5. **Mã nguồn**

```python
import ollama
import gradio as gr
import time

model = "gemma3:4b"

def chatbotResponse(message, history):
    messages = [
        {"role": "system", "content": "Bạn là một giảng viên lớp Tư duy tính toán của trường đại học khoa học tự nhiên, đại học quốc gia thành phố hồ chí minh, việt nam"}
    ]

    if history:
        messages.extend(history)

    text = message.get("text", "")
    image = message.get("files", [])

    if image:
        messages.append({"role": "user", "content": text, "images": image})
    else:
        messages.append({"role": "user", "content": text})

    response = ollama.chat(model=model, messages=messages)
    return response["message"]["content"]

def slow_echo(message, history):
    full_response = chatbotResponse(message, history)

    words = full_response.split(" ")
    partial_response = ""

    for word in words:
        partial_response += word + " "
        time.sleep(0.05)
        yield partial_response

with gr.Blocks() as demo:
    gr.ChatInterface(
        slow_echo,
        save_history=True,
        type="messages",
        multimodal=True,
        title="Ứng dụng Web Chatbot",
        description="<p align='center'> <em>Ứng dụng web chatbot chạy local sử dụng Ollama với model Gemma3:4B và Gradio tạo UI</em> </p>"
    )
    with gr.Sidebar():
        with gr.Accordion("Thông tin sinh viên"):
            gr.Markdown("""
                **Họ và tên**: Nguyễn Anh Thái

                **Mã số sinh viên**: 24120224

                **Học phần**: Tư duy tính toán

                **Nhóm**: CQ2024/2 [N1]
            """)
        with gr.Accordion("Thông tin Model"):
            gr.Markdown("""
                [**Gemma**](https://ollama.com/library/gemma3) is a lightweight, family of models from Google built on Gemini technology. The Gemma 3 models are multimodal—processing text and images—and feature a 128K context window with support for over 140 languages. Available in 270M, 1B, 4B, 12B, and 27B parameter sizes, they excel in tasks like question answering, summarization, and reasoning, while their compact design allows deployment on resource-limited devices.
            """)

if __name__ == "__main__":
    demo.launch()
```
